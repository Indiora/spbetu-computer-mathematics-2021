# Основные теоретические положения

**Бинарный метод**          

Основывается на тождестве, что для любого числа a и чётного числа n выполнимо: $a^n = (a^{n/2})^{n/2} = a^{n/2}*a^{n/2}$. Если n нечетно, то переходим к чётной степени: $a^n = a^{n-1}*a$.                               
Количество операций в бинарном методе можно расчитать по формуле: $\lambda(n) + \nu(n) -1$, где $\lambda(n) = \lfloor lb(n) \rfloor$ и $\nu(n)$ - количество не нулевых цифр в бинарном представлении числа.                                                                                                                                       

**Метод множителей**    

Основывается на тождестве, $x^{ij}$ = $(x^i)^j$, то есть можно отдельно посчитать y = $x^i$, а затем уже вычислять $y^j$. В самом методе $x^n$ разбивается на $(x^i)^j$, где i - наименьший простой делитель n, а j = n / i. В случае если n - просто число, то вычисяется $x^{n-1}$, а результат домножается на x.

**Аддитивная цепочка**

Аддитивная цепочка — последовательность натуральных чисел, начинающаяся с единицы, в которой каждый последующий элемент является суммой каких-то двух предшествующих элементов.  

**Теорема Брауэра**



Для $k < log_2 (log_2 (n))$ справедливо, что: $l(n) < (1+k^{-1}) \cdot \lceil log_2 n \rceil + 2^{k-1} - k + 2$.  

Следствие 1:
Если $\lim \limits_{n \to \infty} \frac{l(n)}{\lambda(n)} = 1$

Следствие 2:
Наилучшая верхняя оценка длины адитивной цепочки равна: $\lambda(n)(1 + \frac{1}{\lambda(\lambda(n))} + \frac{o(\lambda(\lambda(\lambda(n))))}{\lambda(\lambda(n)) ^ 2})$


**Алгоритм Брауэра**

Алгоритм Брауэра задается следующей рекурентной формулой:                                                         
$B_k(n) =\begin{cases}1, 2, 3, ..., 2^k-1\text{, if } n < 2^k \\ B_k(q), 2q, 4q, 8q, ..., 2^kq, n,\text{ if } n \geqslant 2^k\ \text{and } q = \lfloor\frac{n}{2^k}\rfloor \end{cases}$

**Звездная цепочка**

Звездная цепочка - это аддитивная цепочка, в которой текущий элемент получен путем сложения предыдущего и какого-то $a_k$\,  $a_i=a_{i-1}+a_k$ при $0 \leq k \leq i-1$.

**Алгоритм дробления вектора индексов**

1) Во внешнем цикле рассматриваются аддитивные цепочки длины $m$: $\underline{l}(n) \leq m \leq \overline{l}(n)$,  где $\overline{l}(n)$ = $\lambda(n) + \nu(n) -1$ , $\underline{l}(n)$ = $\lceil lb(n) \rceil$

2) Во внутреннем цикле перебираются все фиксированные части $\{r_i\}_{i=1}^{q}$, где $q = \frac{m}{2}$, $q \in N$  на каждом шаге строится $\{r_i\} \to \{a_i\}$, вычисляются $a_{min}$ и $a_{max}$

a) if $a_{m}$ = n $\to $ done

b) if $n \notin [a_{min}, a_{max}] \to $ переходим к следующему набору $\{r_i\}_{i=1}^{q}$

c) if $n \in [a_{min}, a_{max}] \to $ организуем внутренний цикл перебора меняющейся части $\{\rho_i\}_{q+1}^{m}$

d) a) if $a_{m}$ = n $\to $ done

e) if $\rho_i = 1 \ \forall j \ \to $ переходим с слудующему набору $\{r_i\}_{i=1}^{q}$                                  


3) Если $r_i = 1 \ \forall i$ и $\rho_i = 1 \ \forall j \ \to m = m + 1$


**Гипотеза Шольца-Брауэра**                                                                                        
Для минимальных длин звездных цепочек $l^*(n)$ верно $l^*(2^n - 1) <= l^*(n) + n - 1$

**1.Бинарный метод и метод множителей**

```python
# Бинарный метод
def binary_method(x, n):
    # Двоичная запись без последней цифры
    bin_form = bin(n)[3::]
    number_of_operations = 0
    res = x
    for a in str(bin_form):
        # КУ
        if(a == '1'):
            # Умножить на себя
            res = res * res
            # Умножить на x
            res = res * x
            number_of_operations = number_of_operations + 2
        # К
        else:
            # Умножить на себя
            res = res * res
            number_of_operations = number_of_operations + 1

    return [res, number_of_operations]

# Метод множителей
def multiplier_method(x, n):
    # Разложение в произведение простых множителей
    F = factor(n)
    # Если число простое
    if F[0][0] == n:
        # Вызываем функцию с n - 1
        res = multiplier_method(x, n-1)
        # Результат домножаем на х
        res[0] = res[0] * x
        res[1] = res[1] + 1
        return res
    # Если непростое, разбиваем степень на произведение n = i * j
    i = binary_method(x, F[0][0])
    j = binary_method(i[0], n // F[0][0])
    return [j[0], i[1] + j[1]]

print("Бинарный метод при n = 65, x = 1")
print("Кол-во операций: " + str(binary_method(1, 65)[1]))
print("Метод множителей при n = 65, x = 1")
print("Кол-во операций: " + str(multiplier_method(1, 65)[1]) + '\n')

print("Бинарный метод при n = 63, x = 1")
print("Кол-во операций: " + str(binary_method(1, 63)[1]))
print("Метод множителей при n = 63, x = 1")
print("Кол-во операций: " + str(multiplier_method(1, 63)[1]) + '\n')

print("Бинарный метод при n = 67, x = 1")
print("Кол-во операций: " + str(binary_method(1, 67)[1]))
print("Метод множителей при n = 67, x = 1")
print("Кол-во операций: " + str(multiplier_method(1, 67)[1]) + '\n')

print("Бинарный метод при n = 99, x = 1")
print("Кол-во операций: " + str(binary_method(1, 99)[1]))
print("Метод множителей при n = 99, x = 1")
print("Кол-во операций: " + str(multiplier_method(1, 99)[1]) + '\n')

```

**2.Алгоритм Брауэа**

```python
def brauer_algorithm(n, k, additive_chain):
   
    if(n < 2**k):
        # Заполняем цепочку элементами 1, ... , 2^k-1
        [additive_chain.append(i) for i in range(1, 2**k)]
            
    else:
        # Вызываем рекурсивную функцию с  антье(n / 2^k)
        brauer_algorithm( (n // (2**k)), k, additive_chain)
        
        # Добавляем в цепочку 2 * (антье(n / 2^k)), ... , 2^k * (антье(n / 2^k))
        [additive_chain.append(2**i * (n // (2**k))) for i in range(1, k+1)]
            
        # Добавляем в цепочку B(n)
        additive_chain.append(n)
        
def print_result(n, k):
    additive_chain1 = []
    brauer_algorithm(n, k, additive_chain1)
    print("\nПри n = " + str(n) + " и k = " + str(k) + ":")
    print("Длина цепочки = " + str(len(additive_chain1))) 
    print(additive_chain1)


print_result(3141, 4)
print_result(3141, 3)
print_result(3141, 2)
print_result(3141, 1)

print_result(102, 4)
print_result(102, 3)
print_result(102, 2)
print_result(102, 1)
```

Вывод: аддитивные цепочки минимальной длины получаются с помощью алгоритм Брауэра при k =  $k = \lambda(\lambda(n))$, при больших или меньших k длинна цепочки увеличивается, эту закономерность можно увидеть на примере числа 102 выше. Так же алгоритм Брауэра не гарантирует вернуть цепочку наименьшей длины, так например l(102) = 9.

**3.Алгоритм дробления вектора индексов**

```python
import time

# Получаем из вектора индексов аддитивную цепочку
def create_chain(indexV):
    chain = [1]
    for i in indexV:
        chain.append(chain[-1] + chain[i - 1])
    return chain

# Возвращает следующий набор
def create_next_set(vector, q):
    if (vector == [1 for i in range(len(vector))]):
        return [0]
    for i in range(len(vector) - 1, -1, -1):
        if (vector[i] == 1):
            vector[i] = q + i + 1
        else:
            vector[i] -= 1
            return vector
    return vector
    
def split_vector(n):
    # Внешний цикл l_(n) <= m =< _l(n)
    for m in range(int(log(n, 2)), int(log(n, 2)) + bin(n).count('1')):
        q = m // 2 
        r_q = [i for i in range(1, q + 1)]
        r_m = [q + i + 1 for i in range(m - q)]
        # Внутренний цикл, перебо всех наборов r_q (пока не все 1 в векторе)
        while (sum(r_q) >= q ):
            chain = create_chain(r_q + r_m)
            # a_m = n
            if(chain[-1] == n):
                return chain
            # Если n не принадлежит [a_min, a_max]
            if n < (chain[q] + m - q) or chain[q] * 2**(m-q) < n:
                # Переходим к след. набору фикс. часть
                r_q = create_next_set(r_q, 0)
                continue
            # n принадлежит [a_min, a_max], начинаем перебор r_m  (пока не все 1 в векторе)
            while (sum(r_m) >= (m-q)):
                chain = create_chain(r_q + r_m)
                # a_m = n
                if (n == chain[-1]):
                    return chain
                # Переходим к след. набору
                r_m = create_next_set(r_m, q)
            # Переходим к след. набору фикс. часть
            r_q = create_next_set(r_q, 0)
            r_m = [q + i + 1 for i in range(m - q)]
    


def print_result_split(n):
    print('n =', n)
    t1 = time.time()            
    chain = split_vector(n)
    t2 = time.time() - t1
    print(chain)
    print('Длина цепочки:', len(chain))
    print("Время работы алгоримта {0:.5f} s".format(t2), '\n')
    
print_result_split(501)
print_result_split(512)
print_result_split(540)

```

Вывод: Алгоритм дробления вектора индексов всегда находит аддитивную цепочку кратчайшей длинны, в отличии от алгоритма Брауэры или метода множителей, но так как алгоритм переборный время работы неравномерно для разных n и может быть очень большим.
