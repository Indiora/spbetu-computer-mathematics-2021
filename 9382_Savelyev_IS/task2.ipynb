Задание: Сколькими способами можно составить бусы из 161 бусинки 6 различных цветов(с учетом поворотов и осевых семметрий) ?

Дано два множества, первое множество бусинок $\{x_1, x_2,...,x_{161}\}$, второе множество цветов  $\{a,b,c,d,e,f\}$. Каждая бусинка окрашена в один из 6 цветов, всего вариантов раскрасить бусы $6^{161}$. Но по условию бусы полученные с помощью операции поворота или осевой симметрии, считаются одинаковыми.

Таким образом на множестве всех раскрасок $X$ действует группа поворотов и осевых симметрий. Операция в этой группе - композиция, то есть последовательное выполнение преобразований. Операция композиции является ассоциативнй, нейтральным элементов является тождественное отображение, обратным к осевой симметрии она сама, а к повороту, поворот в другую сторону. Обозначим эту группу буквой $G$.

Группа $G$ действует на множество мест $\{x_1, x_2,...,x_{161}\}$, то есть задано отображение из $G$ в симметрическую группу $S_{161}$. Обозначим через $\tau$ поворот, переводящий каждое место в следующее, а через $\sigma$ обозначим осевую симметрию.

...

$\sigma \tau^{161} = \sigma = (1,161)(2,160)(3,159)...(79,83)(80,82)(81)$

$\sigma^2 = e$
